```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
sqlitedb_path = '/home/usuario/Documentos/Ironhack/Ironhack-Module-1-Project---Commercial-Optimization/data/raw/juankmp.db'
engine = create_engine(f'sqlite:///{sqlitedb_path}')

df_all_rich_origin_raw = pd.read_sql_query("SELECT * from personal_info pi left join business_info bi on pi.id = bi.id left join rank_info ri on pi.id = ri.id", engine)
df_all_rich_origin_raw.head()

df_all_rich_origin = pd.read_sql_query("select ri.id,ri.name,ri.position,pi.lastName,pi.age,pi.gender,pi.country,pi.image,bi.Source,bi.worth,bi.worthChange,bi.realTimeWorth,bi.realTimePosition from rank_info ri left join personal_info pi on ri.id = pi.id left join business_info bi on ri.id = bi.id", engine)
df_all_rich_origin.head()

df_perinfo_origin = pd.read_sql_query("SELECT * from personal_info pi", engine)
display(df_perinfo_origin.head())

df_businfo_origin = pd.read_sql_query("SELECT * from business_info bi", engine)
display(df_businfo_origin.head())

df_rankinfo_origin = pd.read_sql_query("SELECT * from rank_info pi", engine)
display(df_rankinfo_origin.head())
```
