# Data Cleaning!


```python
import pandas as pd
```

```python
#df_per_info = pd.read_csv('/home/usuario/Documentos/Ironhack/Ironhack-Module-1-Project---Commercial-Optimization/data/raw/perinfo_origin.csv')
#df_bus_info = pd.read_csv('/home/usuario/Documentos/Ironhack/Ironhack-Module-1-Project---Commercial-Optimization/data/raw/businfo_origin.csv')
#df_ran_info = pd.read_csv('/home/usuario/Documentos/Ironhack/Ironhack-Module-1-Project---Commercial-Optimization/data/raw/rankinfo_origin.csv')
df_allrich_info = pd.read_csv('/home/usuario/Documentos/Ironhack/Ironhack-Module-1-Project---Commercial-Optimization/data/raw/allrich_origin.csv')
df_allrich_info.head()
```

```python
#converting null in zero
def conv_null_zero(x):
        if pd.notna(x) :
            return x
        else:
            return 0
```

```python
#1 Last_name treatment = Capitalize all letters & null values in not given.

def capitalize(x):
    try:
        return x.capitalize()
    except:
        return x

def conv_not_name(x):
        if pd.notna(x):
            return x
        else:
            return 'not_name_given'

df_allrich_info['lastName'] = df_allrich_info['lastName'].apply(capitalize) 
df_allrich_info['lastName'] = df_allrich_info['lastName'].apply(conv_not_name) 

df_allrich_info['name'] = df_allrich_info['name'].apply(capitalize) 
df_allrich_info['name'] = df_allrich_info['name'].apply(conv_not_name)
```

```python
#2 age=years (4 digits) calculation to actual age.

def age_calc(x):
    try:
        if len(str(x)) == 4:
            new_age = 2019 - int(x)
            return int(new_age)
        else:
            return int(x)
    except:
        return x

#2 age     
df_allrich_info['age'] = df_allrich_info['age'].apply(age_calc)

```

```python
#2.1  The rest of the figures remove "years old"

def replace_str(x):
    try:
        return int(x.replace(' years old',""))
    except:
        return x

    
#2.1 age/years old
df_allrich_info['age'] = df_allrich_info['age'].apply(replace_str)

#Convert null years in zero
df_allrich_info['age'] = df_allrich_info['age'].apply(conv_null_zero) 
    
```

```python
#5 worth = removed BUSD

def convert_billions(x):
    try:
        return float(x.replace(' BUSD',""))
    except:
        return float(x)

df_allrich_info['worth'] = df_allrich_info['worth'].apply(convert_billions)
df_allrich_info['worth'] = df_allrich_info['worth'].apply(conv_null_zero)

#rename column with value dimension US from worth to worth_in_BUSD
df_allrich_info.rename(columns={"worth": "worth_in_BUSD"}, inplace = True)

```

```python
#6 worthchange = removed millions USD and then multiply it for 1.000.000

def convert_millions(x):
    try:
        return float(x.replace(' millions USD',""))
    except:
        return float(x)

df_allrich_info['worthChange'] = df_allrich_info['worthChange'].apply(convert_millions)
df_allrich_info['worthChange'] = df_allrich_info['worthChange'].apply(conv_null_zero)

#rename column with value dimension US from worthChange to worthChange_in_USD
df_allrich_info.rename(columns={"worthChange": "worthChange_in_MSD"}, inplace = True)

```

```python
#4 source --> Divided by 2 columns from ==> first one is industry_1 second industry_2 
df_allrich_info[['Industry','Company']] = df_allrich_info["Source"].str.split("==> ",expand=True,)

#eliminate source column
df_allrich_info = df_allrich_info.drop('Source', 1)

df_allrich_info['Industry'] = df_allrich_info['Industry'].apply(capitalize) 
df_allrich_info['Company'] = df_allrich_info['Company'].apply(capitalize) 
```

```python
#converting null country in not_country_given
          
def conv_not_country(x):
        if pd.notna(x) :
            return x
        else:
            return 'not_country_given'
        
df_allrich_info['country'] = df_allrich_info['country'].apply(conv_not_country)
df_allrich_info['country']=df_allrich_info['country'].replace('None','not_country_given')

df_allrich_info['country']=df_allrich_info['country'].replace('USA','United States')
df_allrich_info['country']=df_allrich_info['country'].replace('UK','United Kingdom')
df_allrich_info['country']=df_allrich_info['country'].replace("People's Republic of China",'China')

```

```python
#converting null in zero
def conv_gender(x):
        if x == 'Male':
            return 'M'
        elif x == 'Female':
            return 'F'
        else:
            return x

def conv_not_gender(x):
        if pd.notna(x):
            return x
        else:
            return 'not_gender_given'
```

```python
df_allrich_info['gender'] = df_allrich_info['gender'].apply(conv_gender)
df_allrich_info['gender'] = df_allrich_info['gender'].apply(conv_not_gender)
df_allrich_info['gender']=df_allrich_info['gender'].replace('None','not_gender_given')
display(df_allrich_info['gender'])
```
